```python
deliveries["date"] = pd.to_datetime(deliveries["date"])

deliveries["date"].head()
```

```python
deliveries["total_runs"] = deliveries["runs_of_bat"] + deliveries["extras"]
```

```python
deliveries["legal_ball"] = (deliveries["wide"] == 0).astype(int)
```

```python
deliveries["is_four"] = (deliveries["runs_of_bat"] == 4).astype(int)

deliveries["is_six"] = (deliveries["runs_of_bat"] == 6).astype(int)

deliveries["is_boundary"] = (
    (deliveries["runs_of_bat"] == 4) |
    (deliveries["runs_of_bat"] == 6)
).astype(int)
```

```python
deliveries["is_dot"] = (
    deliveries["total_runs"] == 0
).astype(int)
```

```python
deliveries["is_wicket"] = (
    deliveries["player_dismissed"].notna()
).astype(int)
```

```python
deliveries["over_phase"] = np.select(
    [
        deliveries["over"] <= 6,
        deliveries["over"] <= 15,
        deliveries["over"] <= 20
    ],
    [
        "Powerplay",
        "Middle Overs",
        "Death Overs"
    ],
    default="Unknown"
)
```

```python
# Calculate total runs scored by each batter
top_batters = (
    clean_deliveries
    .groupby("striker")["runs_of_bat"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_batters
```

```python
# Calculate batting statistics
batting_stats = (
    clean_deliveries
    .groupby("striker")
    .agg(
        Runs=("runs_of_bat", "sum"),
        Balls=("legal_ball", "sum")
    )
)

# Keep batters with at least 100 balls
batting_stats = batting_stats[batting_stats["Balls"] >= 100]

# Calculate strike rate
batting_stats["Strike Rate"] = (
    batting_stats["Runs"] / batting_stats["Balls"] * 100
)

# Sort by strike rate
top_sr = (
    batting_stats
    .sort_values("Strike Rate", ascending=False)
    .head(10)
)

top_sr
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    x=top_sr.index,
    y=top_sr["Strike Rate"]
)

plt.title("Top 10 Strike Rates (Min. 100 Balls)")
plt.xlabel("Batter")
plt.ylabel("Strike Rate")

plt.xticks(rotation=45)

plt.show()
```

```python
# Calculate batting statistics for each batter
batter_stats = (
    clean_deliveries
    .groupby("striker")
    .agg(
        Runs=("runs_of_bat", "sum"),
        Balls=("legal_ball", "sum"),
        Fours=("is_four", "sum"),
        Sixes=("is_six", "sum"),
        Boundaries=("is_boundary", "sum"),
        Dot_Balls=("is_dot", "sum"),
        Dismissals=("is_wicket", "sum")
    )
)

# Calculate strike rate
batter_stats["Strike_Rate"] = (
    batter_stats["Runs"] / batter_stats["Balls"] * 100
).round(2)

# Calculate batting average
batter_stats["Average"] = (
    batter_stats["Runs"] / batter_stats["Dismissals"].replace(0, np.nan)
).round(2)

# Calculate boundary percentage
batter_stats["Boundary_%"] = (
    batter_stats["Boundaries"] / batter_stats["Balls"] * 100
).round(2)

# Calculate dot ball percentage
batter_stats["Dot_Ball_%"] = (
    batter_stats["Dot_Balls"] / batter_stats["Balls"] * 100
).round(2)

batter_stats.head(10)
```

```python
top_runs = batter_stats.sort_values("Runs", ascending=False).head(10)

top_runs
```

```python
# Top batters by boundary percentage (minimum 100 balls)
top_boundary = (
    batter_stats[batter_stats["Balls"] >= 100]
    .sort_values("Boundary_%", ascending=False)
    .head(10)
)

top_boundary
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    data=top_boundary,
    x=top_boundary.index,
    y="Boundary_%"
)

plt.title("Top Boundary Percentage (Min. 100 Balls)")
plt.xlabel("Batter")
plt.ylabel("Boundary %")

plt.xticks(rotation=45)
plt.show()
```

```python
# Batters with the lowest dot ball percentage
lowest_dot = (
    batter_stats[batter_stats["Balls"] >= 100]
    .sort_values("Dot_Ball_%")
    .head(10)
)

lowest_dot
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    data=lowest_dot,
    x=lowest_dot.index,
    y="Dot_Ball_%"
)

plt.title("Lowest Dot Ball Percentage (Min. 100 Balls)")
plt.xlabel("Batter")
plt.ylabel("Dot Ball %")

plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate bowling statistics for each bowler
bowler_stats = (
    clean_deliveries
    .groupby("bowler")
    .agg(
        Balls=("legal_ball", "sum"),
        Runs_Conceded=("total_runs", "sum"),
        Wickets=("is_wicket", "sum"),
        Dot_Balls=("is_dot", "sum")
    )
)

# Calculate economy rate
bowler_stats["Economy"] = (
    bowler_stats["Runs_Conceded"] / (bowler_stats["Balls"] / 6)
).round(2)

# Calculate bowling strike rate
bowler_stats["Strike_Rate"] = (
    bowler_stats["Balls"] / bowler_stats["Wickets"].replace(0, np.nan)
).round(2)

# Calculate bowling average
bowler_stats["Average"] = (
    bowler_stats["Runs_Conceded"] / bowler_stats["Wickets"].replace(0, np.nan)
).round(2)

# Calculate dot ball percentage
bowler_stats["Dot_Ball_%"] = (
    bowler_stats["Dot_Balls"] / bowler_stats["Balls"] * 100
).round(2)

bowler_stats.head()
```

```python
# Calculate average score at each venue
avg_score = (
    clean_deliveries
    .groupby(["venue", "match_id", "innings"])["total_runs"]
    .sum()
    .groupby("venue")
    .mean()
    .rename("Average_Score")
)

# Calculate total runs at each venue
total_runs = (
    clean_deliveries
    .groupby("venue")["total_runs"]
    .sum()
    .rename("Total_Runs")
)

# Calculate total boundaries
boundaries = (
    clean_deliveries
    .groupby("venue")["is_boundary"]
    .sum()
    .rename("Boundaries")
)

# Calculate total wickets
wickets = (
    clean_deliveries
    .groupby("venue")["is_wicket"]
    .sum()
    .rename("Wickets")
)

# Calculate legal balls
balls = (
    clean_deliveries
    .groupby("venue")["legal_ball"]
    .sum()
)

# Calculate run rate
run_rate = (
    total_runs / (balls / 6)
).rename("Run_Rate")

# Combine all metrics into one table
venue_summary = pd.concat(
    [
        avg_score,
        total_runs,
        boundaries,
        wickets,
        run_rate
    ],
    axis=1
).round(2)

# Sort by average score
venue_summary = venue_summary.sort_values(
    by="Average_Score",
    ascending=False
)

venue_summary
```

```python
# Count matches hosted at each venue
matches = (
    clean_deliveries
    .groupby("venue")["match_id"]
    .nunique()
    .rename("Matches")
)

venue_summary = venue_summary.join(matches)

venue_summary = venue_summary[
    [
        "Matches",
        "Average_Score",
        "Run_Rate",
        "Total_Runs",
        "Boundaries",
        "Wickets"
    ]
]

venue_summary
```

```python
# Save the summary table
venue_summary.to_csv("venue_summary.csv", index=True)
```
